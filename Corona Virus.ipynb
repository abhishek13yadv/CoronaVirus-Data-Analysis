```python
all_country_recovery = {}
all_country_death = {}
for country in virus['Country']:
    if country not in all_country_recovery and country != 'Mainland China':
        all_country_recovery[country] = virus.loc[virus['Country']==country,'Recovered'].sum()
        all_country_death[country] = virus.loc[virus['Country']==country,'Deaths'].sum()
```

```python
all_country_recovery.values()
```

```python
all_country_death.values()
```

```python
plt.scatter(all_country_death.values(),all_country_recovery.values())
plt.xlabel('Deaths a/c country')
plt.ylabel('Recovery a/c country')
plt.show()
```

### I tried creating a scatter plot between the recovery and death cases for different countries. But it does not seem possible because of Maniland China being an extraordinary outlier. So I removed that country and plotted it.

```python
all_province_recovery = {}
all_province_death = {}
for province in virus['Province']:
    if province not in all_province_recovery and province != 'Hubei':
        all_province_recovery[province] = virus.loc[virus['Province']==province,'Recovered'].sum()
        all_province_death[province] = virus.loc[virus['Province']==province,'Deaths'].sum()
```

```python
all_province_recovery
```

```python
plt.scatter(all_province_death.values(),all_province_recovery.values())
plt.xlabel('Deaths a/c province')
plt.ylabel('Recovery a/c province')
plt.show()
```
